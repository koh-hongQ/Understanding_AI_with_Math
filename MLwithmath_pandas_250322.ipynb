# 과제

1.   결측치 처리(Age,Cabin,Embarked)
2.   범주형데이터 처리(원핫인코딩:Sex, Embarked)
3. PassengerId, Name, Ticket 행 삭제


```python
#데이터 불러오기
train=pd.read_csv("drive/My Drive/Colab Notebooks/2025MLwithmath/titanic/train.csv")
```

```python
#전처리 함수
def preprocessing(df):
  df['Age']=df['Age'].fillna(df['Age'].mean())
  df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
  df['Cabin'] = df['Cabin'].fillna(df['Cabin'].mode()[0])
  #원핫인코딩
  df_embarked=pd.get_dummies(df['Embarked'])
  df=pd.concat([df,df_embarked],axis=1)
  df=df.drop('Embarked',axis=1)
  df_sex=pd.get_dummies(df['Sex'])
  df=pd.concat([df,df_sex],axis=1)
  df=df.drop('Sex',axis=1)

  df = df.drop('PassengerId', axis=1)
  df = df.drop('Name', axis=1)
  df = df.drop('Ticket', axis=1)
  return df
```

```python
train_pre = preprocessing(train)
train_pre.head()
```
